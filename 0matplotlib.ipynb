# Matplotlib

```python
# import
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Basic features
x = list(range(10))
y = list(range(9,-1,-1))
x, y
```

```python
df = pd.DataFrame({'x': x, 'y': y})
df.head(3)
```

```python
# plt.plot(x, y)
plt.plot(df['x'], df['y'])
plt.title("Grafik x vs y", fontdict={
    'fontsize':15, 'family': 'Impact', 'color': 'red'}
)
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.grid(True)
plt.legend(['Dataku'], loc=0)
plt.show() # mandatory for python script, optional for notebook 
```

```python
# line style
# lines:  -  --  :  -.
plt.plot(df['x'], df['y'], '-')
plt.plot(df['x'], [0]*10, '--')
plt.plot([0]*10, df['y'], ':')
plt.plot([9]*10, df['y'], '-.')
```

```python
# marker style
# marker: o s * ^ . + D
plt.plot(df['x'], df['y'], 'o')
plt.plot(df['x'], [0]*10, 's')
plt.plot([0]*10, df['y'], '^')
plt.plot([9]*10, df['y'], '*')
```

```python
# line & marker
# warna:  g green  r red  b blue  y yellow  k black
plt.plot(df['x'], df['y'], 'go--')
plt.plot(df['x'], [0]*10, '--or')
plt.plot([0]*10, df['y'], 'yD:')
plt.plot([9]*10, df['y'], '-*k')
```

```python
# save as image
plt.plot(df['x'], df['y'])
plt.savefig('0.png')
```

## 2. Bar Chart

```python
plt.bar(x, y)
for i in x:
    plt.text(i-.1, y[i]+.2, y[i])
```

```python
for i in list(df['Type 1'].unique()):
    plt.bar(i, df[df['Type 1'] == i].count()['Name'], label=10)
    plt.xticks(rotation=90)
    plt.text(
        list(df['Type 1'].unique()).index(i) - .3,  # posisi x
        df[df['Type 1'] == i].count()['Name'] + 1, # posisi y
        df[df['Type 1'] == i].count()['Name']  # teks yang ditampilkan
    )
```

```python
# horizontal bar chart
for i in list(df['Type 1'].unique()):
    plt.barh(i, df[df['Type 1'] == i].count()['Name'], label=10)
    plt.text(
        df[df['Type 1'] == i].count()['Name'] + 1,  # posisi x
        list(df['Type 1'].unique()).index(i) - .3,   # posisi y
        df[df['Type 1'] == i].count()['Name']        # teks yang ditampilkan
    )
```

## 4. Pie Chart

```python
pendidikan = ['SD', 'SMP', 'SMA']
jumlah = [10, 14, 5]
wedge, label, percentage = plt.pie(jumlah, labels=pendidikan, autopct='%1.1f%%') 
for i in percentage:
    i.set_color('white')
for i in label:
    i.set_color('red')
```

```python
type = list(df['Type 1'].unique())
jumlah = []
for i in type:
    jumlah.append(df[df['Type 1'] == i].count()['Name'])

wedge, label, percentage = plt.pie(jumlah, labels=type, autopct='%1.1f%%') 
for i in percentage:
    i.set_color('white')
for i in label:
    i.set_color('red')
```

```python
import folium
```

```python
lat = -6.9053905
long = 106.8736239
m = folium.Map(
    location = [lat, long],
    tiles = 'OpenStreetMap',
    zoom_start = 15
)
folium.Marker(
    [lat, long],
    popup = 'Ini Nusa Putra',
    tooltip = 'Aku di sini',
    icon = folium.Icon(color='purple', icon='info-sign')
).add_to(m)
m
```
